```python
import pandas as pd
a = pd.read_csv('RMSDnew.csv')
a
```

CREATING A NEW COLUMN


```python
a['frame']=  a['frame#'] / 10
a
```

PLOTTING RMSD

```python

import matplotlib.pyplot as plt
import sys

plt.figure(figsize=(10,10))
plt.plot(a['frame'], a['Enalapril RMSD (prot BacKbone)'], color='yellow')
plt.plot(a['frame'], a['Isovitexin Prot_Backbone'], color='blue')   
plt.plot(a['frame'], a['Ngenin-7-O-glucoProt_Backbone'], color='green')
plt.plot(a['frame'], a['ApoproteinProt_Backbone'], color='red')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (Ã…)')
plt.title('RMSD')
plt.legend(['Enalapril RMSD (prot BacKbone)_ACE', 'Isovitexin Prot_Backbone_ACE', 'Ngenin-7-O-glucoProt_Backbone_ACE', 'ApoproteinProt_Backbone_ACE'], loc='upper left')




plt.show()




```
