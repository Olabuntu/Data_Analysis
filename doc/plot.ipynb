```python
import pandas as pd
import matplotlib.pyplot as plt

h1 = pd.read_csv('hbonds_Genistein.csv')
h2 = pd.read_csv('hbonds_Apo.csv')
h3 = pd.read_csv('hbonds_MUP.csv')
h4 = pd.read_csv('hbonds_Sinapic.csv')

rd1 = pd.read_csv('RMSD_Genistein.csv')
rd2 = pd.read_csv('RMSD_Apo.csv')
rd3 = pd.read_csv('RMSD_MUP.csv')
rd4 = pd.read_csv('RMSD_Sinapic.csv')

rf1 = pd.read_csv('RMSF_Genistein.csv')
rf2 = pd.read_csv('RMSF_Apo.csv')
rf3 = pd.read_csv('RMSF_MUP.csv')
rf4 = pd.read_csv('RMSF_Sinapic.csv')

rog1 = pd.read_csv('radius_of_gyration_Genistein.csv')
rog2 = pd.read_csv('radius_of_gyration_Apo.csv')
rog3 = pd.read_csv('radius_of_gyration_MUP.csv')
rog4 = pd.read_csv('radius_of_gyration_Sinapic.csv')

s1 = pd.read_csv('SASA_protein_Genistein.csv')
s2 = pd.read_csv('SASA_protein_Apo.csv')
s3 = pd.read_csv('SASA_protein_MUP.csv')
s4 = pd.read_csv('SASA_protein_Sinapic.csv')


H1_Genistein=h1['H']
H2_Apo=h2['H']
H3_MUP=h3['H']
H4_Sinapic=h4['H']
h1['time']=h1.index / 10
H5_time=h1['time']



RD1_Genistein=rd1['RMSD']
RD2_Apo=rd2['RMSD']
RD3_MUP=rd3['RMSD']
RD4_Sinapic=rd4['RMSD']
rd1['time']= rd1.index / 10
RD5_time=rd1['time']



RF1_Genistein=rf1['RMSF']
RF2_Apo=rf2['RMSF']
RF3_MUP=rf3['RMSF']
RF4_Sinapic=rf4['RMSF']
RF5_AA=rf1['A']



ROG1_Genistein=rog1['ROG']
ROG2_Apo=rog2['ROG']
ROG3_MUP=rog3['ROG']
ROG4_Sinapic=rog4['ROG']
rog1['time']=rog1.index / 10
ROG_time = rog1['time']



S1_Genistein=s1['SASA']
S2_Apo=s2['SASA']
S3_MUP=s3['SASA']
S4_Sinapic=s4['SASA']
s1['time']=s1.index / 10
S5_time=s1['time']



hbond = pd.concat([H5_time, H2_Apo, H3_MUP, H1_Genistein, H4_Sinapic], axis=1)
hbond.columns = ['H5_time', 'H2_Apo', 'H3_MUP', 'H1_Genistein', 'H4_Sinapic']

rmsd = pd.concat([RD5_time, RD2_Apo, RD3_MUP, RD1_Genistein, RD4_Sinapic], axis=1)
rmsd.columns = ['RD5_time', 'RD2_Apo', 'RD3_MUP', 'RD1_Genistein', 'RD4_Sinapic']

rmsf = pd.concat([RF5_AA, RF2_Apo, RF3_MUP, RF1_Genistein, RF4_Sinapic], axis=1)
rmsf.columns = ['RF5_AA', 'RF2_Apo', 'RF3_MUP', 'RF1_Genistein', 'RF4_Sinapic']

rog = pd.concat([ROG_time, ROG2_Apo, ROG3_MUP, ROG1_Genistein, ROG4_Sinapic], axis=1)
rog.columns = ['ROG_time', 'ROG2_Apo', 'ROG3_MUP', 'ROG1_Genistein', 'ROG4_Sinapic']

sasa = pd.concat([S5_time, S2_Apo, S3_MUP, S1_Genistein, S4_Sinapic], axis=1)
sasa.columns = ['S5_time', 'S2_Apo', 'S3_MUP', 'S1_Genistein', 'S4_Sinapic']

fig, ax = plt.subplots(2,2,figsize=(15,10))

ax[1,0].plot(hbond['H5_time'],hbond['H2_Apo'], color = 'blue')
ax[1,0].plot(hbond['H5_time'],hbond['H3_MUP'], color = 'green')
ax[1,0].plot(hbond['H5_time'],hbond['H1_Genistein'], color = 'red')
ax[1,0].plot(hbond['H5_time'],hbond['H4_Sinapic'], color = 'black')
ax[1,0].set_xlabel('Time (ns)')
ax[1,0].set_ylabel('Number of H-bonds')
ax[1,0].set_title('H-bonds')
ax[1,0].set_title('C' , loc='left')



ax[0,0].plot(rmsd['RD5_time'],rmsd['RD2_Apo'], color = 'blue')
ax[0,0].plot(rmsd['RD5_time'],rmsd['RD3_MUP'], color = 'green')
ax[0,0].plot(rmsd['RD5_time'],rmsd['RD1_Genistein'], color = 'red')
ax[0,0].plot(rmsd['RD5_time'],rmsd['RD4_Sinapic'], color = 'black')
ax[0,0].set_xlabel('Time (ns)')
ax[0,0].set_ylabel('Distance ()')
ax[0,0].set_title('RMSD')
ax[0,0].set_title('A' , loc='left')


ax[0,1].plot(rog['ROG_time'],rog['ROG2_Apo'], color = 'blue')
ax[0,1].plot(rog['ROG_time'],rog['ROG3_MUP'], color = 'green')
ax[0,1].plot(rog['ROG_time'],rog['ROG1_Genistein'], color = 'red')
ax[0,1].plot(rog['ROG_time'],rog['ROG4_Sinapic'], color = 'black')
ax[0,1].set_xlabel('Time (ns)')
ax[0,1].set_ylabel('Distance ()')
ax[0,1].set_title('ROG')
ax[0,1].set_title('B' , loc='left')


ax[1,1].plot(sasa['S5_time'],sasa['S2_Apo'], color = 'blue')
ax[1,1].plot(sasa['S5_time'],sasa['S3_MUP'], color = 'green')
ax[1,1].plot(sasa['S5_time'],sasa['S1_Genistein'], color = 'red')
ax[1,1].plot(sasa['S5_time'],sasa['S4_Sinapic'], color = 'black')
ax[1,1].set_xlabel('Time (ns)')
ax[1,1].set_ylabel('Area ()')
ax[1,1].set_title('SASA')
ax[1,1].set_title('D' , loc='left')

fig.legend(['Apoprotein','MUP','Genistein','Sinapic'],loc= 'lower center', ncol=4)
plt.show()






```
